```python
# API: STATION AVAILABILITY:
import requests
import pandas as pd

url = 'https://opendata.emel.pt/cycling/gira/station/availability'

response = requests.get(url)

st_availability = response.json()
```

```python
### Examplo of hoy to get number of bikes available:
st_availability['features'][0]['properties']['num_bicicletas']
```

```python
### Function that get the number of available bikes:

def get_bikes_availables(station_id):
    station_data = next((station['properties'] for station in st_availability['features'] if station['properties']['id_expl'] == str(station_id)), None)
    return station_data['num_bicicletas']
```

```python
get_bikes_availables(419)
```

```python
### Code to build dataframe with relevant data:

data_list = []

for i in st_availability['features']:
    name = i['properties']
    data_list.append({'id_expl': name['id_expl'], 'desig_comercial': name['desig_comercial'], 'num_bicicletas': name['num_bicicletas'], 'num_docas': name['num_docas'],'update_date':name['update_date']})

# Create DataFrame
list_id_station = pd.DataFrame(data_list)
list_id_station
```

```python
list_id_station.to_json('list_id_station.json', orient='records', lines=True)
```

```python
# API: STATISTICS USAGE:

url = 'https://opendata.emel.pt/cycling/gira/statistics/usage'

response = requests.get(url)

statistics = response.json()

df_statistics = pd.DataFrame(statistics)
```

```python
df_statistics
```

```python
# Sorting values by dates

df_statistics['tripStartDate'] = pd.to_datetime(df_statistics['tripStartDate']) 
df_statistics = df_statistics.sort_values(by='tripStartDate', ignore_index=True)
```

```python
df_statistics.to_json('df_statistics.json', orient='records', lines=True)
```
